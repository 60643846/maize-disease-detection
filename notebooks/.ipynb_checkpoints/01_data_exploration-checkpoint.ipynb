```python
# CELL: Convert YOLO Test Data to Classification Format (Fixed)

import shutil

TEST_DIR = PROCESSED_DIR / "test"
IMAGES_DIR = TEST_DIR / "images"
LABELS_DIR = TEST_DIR / "labels"

print("Converting YOLO format to classification folders...")
print("="*70)

# Create class folders
MLB_DIR = TEST_DIR / "MLB"
MSV_DIR = TEST_DIR / "MSV"
MLB_DIR.mkdir(exist_ok=True)
MSV_DIR.mkdir(exist_ok=True)

# Class mapping from data.yaml
class_names = {0: 'MLB', 1: 'MSV'}

# Process each label file
label_files = list(LABELS_DIR.glob("*.txt"))
print(f"Found {len(label_files)} label files\n")

mlb_count = 0
msv_count = 0
skipped = 0
errors = 0

for label_file in label_files:
    try:
        # Check if label file actually exists
        if not label_file.exists():
            skipped += 1
            continue
            
        # Read class from label file
        with open(label_file, 'r') as f:
            line = f.readline().strip()
            if not line:
                skipped += 1
                continue
                
            class_id = int(line.split()[0])  # First number is class
            
            # Find corresponding image
            img_name = label_file.stem + ".jpg"
            img_path = IMAGES_DIR / img_name
            
            # Check if image exists
            if not img_path.exists():
                print(f"Warning: Image not found for {label_file.name}")
                skipped += 1
                continue
            
            # Copy to appropriate class folder
            dest_path = None
            if class_id == 0:
                dest_path = MLB_DIR / img_name
                if not dest_path.exists():  # Don't copy if already exists
                    shutil.copy(img_path, dest_path)
                    mlb_count += 1
            elif class_id == 1:
                dest_path = MSV_DIR / img_name
                if not dest_path.exists():
                    shutil.copy(img_path, dest_path)
                    msv_count += 1
                    
    except Exception as e:
        print(f"Error processing {label_file.name}: {e}")
        errors += 1
        continue

print("\nConversion complete!")
print("="*70)
print(f"MLB (class 0): {mlb_count} new images copied")
print(f"MSV (class 1): {msv_count} new images copied")
print(f"Skipped: {skipped}")
print(f"Errors: {errors}")

# Count total in each folder
total_mlb = len(list(MLB_DIR.glob("*.jpg")))
total_msv = len(list(MSV_DIR.glob("*.jpg")))
print(f"\nTotal in MLB folder: {total_mlb}")
print(f"Total in MSV folder: {total_msv}")
print(f"Grand total: {total_mlb + total_msv}")
print("="*70)

# Ask before cleaning up
print("\nYOLO folders (images/ and labels/) still exist.")
print("Run next cell to clean them up after verifying the conversion worked.")
```

```python
# CELL: Define Image Quality Analysis Function (Fixed)

import cv2
import numpy as np
from pathlib import Path
import random

def analyze_image_quality(data_dir, sample_size=100):
    """
    Analyze image quality metrics (brightness, sharpness, resolution)
    """
    # Collect all image paths
    all_images = []
    
    # Check if data_dir itself contains images (single class folder)
    direct_images = list(data_dir.glob("*.jpg")) + list(data_dir.glob("*.JPG"))
    
    if direct_images:
        # Images are directly in this folder
        all_images = direct_images
    else:
        # Images are in subdirectories (multi-class folder)
        for class_dir in data_dir.iterdir():
            if class_dir.is_dir():
                all_images.extend(list(class_dir.glob("*.jpg")) + list(class_dir.glob("*.JPG")))
    
    if not all_images:
        print(f"ERROR: No images found in {data_dir}")
        return [], [], [], []
    
    # Sample images if we have more than sample_size
    if len(all_images) > sample_size:
        all_images = random.sample(all_images, sample_size)
    
    brightness_values = []
    sharpness_values = []
    resolutions = []
    
    for img_path in all_images:
        img = cv2.imread(str(img_path))
        if img is not None:
            # Brightness (average pixel intensity)
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            brightness = np.mean(gray)
            brightness_values.append(brightness)
            
            # Sharpness (Laplacian variance)
            laplacian = cv2.Laplacian(gray, cv2.CV_64F)
            sharpness = laplacian.var()
            sharpness_values.append(sharpness)
            
            # Resolution
            height, width = img.shape[:2]
            resolutions.append((width, height))
    
    # Print summary
    if brightness_values:
        print(f"Analyzed {len(brightness_values)} images")
        print(f"Average brightness: {np.mean(brightness_values):.1f} (range: {np.min(brightness_values):.1f} - {np.max(brightness_values):.1f})")
        print(f"Average sharpness: {np.mean(sharpness_values):.1f} (range: {np.min(sharpness_values):.1f} - {np.max(sharpness_values):.1f})")
    else:
        print("No valid images analyzed")
    
    return brightness_values, sharpness_values, resolutions, all_images

print("Image quality analysis function defined (fixed for single-class folders)")
```

```python
# CELL: Compare clean vs noisy image quality

print("\n" + "="*70)
print("COMPARING IMAGE QUALITY: CLEAN vs NOISY")
print("="*70 + "\n")

# Analyze PlantVillage (clean) - sample from test_clean
CLEAN_TEST = PROCESSED_DIR / "test_clean"
print("PLANTVILLAGE (CLEAN) - Baseline:")
print("-"*70)
clean_brightness, clean_sharpness, _, _ = analyze_image_quality(CLEAN_TEST, sample_size=100)
print("\n")

# Analyze Maize MLB (noisy)
MLB_NOISY = PROCESSED_DIR / "test" / "MLB"
print("MAIZE MLB (NOISY) - Real-world:")
print("-"*70)
mlb_brightness, mlb_sharpness, _, _ = analyze_image_quality(MLB_NOISY, sample_size=100)
print("\n")

# Analyze Maize MSV (noisy)
MSV_NOISY = PROCESSED_DIR / "test" / "MSV"
print("MAIZE MSV (NOISY) - Real-world:")
print("-"*70)
msv_brightness, msv_sharpness, _, _ = analyze_image_quality(MSV_NOISY, sample_size=100)

# Summary comparison
print("\n" + "="*70)
print("SUMMARY COMPARISON:")
print("="*70)
print(f"{'Dataset':<25} {'Brightness':>15} {'Sharpness':>15} {'Quality':>15}")
print("-"*70)
print(f"{'PlantVillage (clean)':<25} {np.mean(clean_brightness):>15.1f} {np.mean(clean_sharpness):>15.1f} {'Baseline':>15}")
print(f"{'Maize MLB (noisy)':<25} {np.mean(mlb_brightness):>15.1f} {np.mean(mlb_sharpness):>15.1f} {'Test':>15}")
print(f"{'Maize MSV (noisy)':<25} {np.mean(msv_brightness):>15.1f} {np.mean(msv_sharpness):>15.1f} {'Test':>15}")
print("="*70)

# Calculate differences
brightness_drop_mlb = np.mean(clean_brightness) - np.mean(mlb_brightness)
brightness_drop_msv = np.mean(clean_brightness) - np.mean(msv_brightness)
sharpness_drop_mlb = ((np.mean(clean_sharpness) - np.mean(mlb_sharpness)) / np.mean(clean_sharpness)) * 100
sharpness_drop_msv = ((np.mean(clean_sharpness) - np.mean(msv_sharpness)) / np.mean(clean_sharpness)) * 100

print(f"\nINSIGHTS:")
print(f"  MLB brightness drop: {brightness_drop_mlb:.1f} points ({brightness_drop_mlb/np.mean(clean_brightness)*100:.1f}%)")
print(f"  MSV brightness drop: {brightness_drop_msv:.1f} points ({brightness_drop_msv/np.mean(clean_brightness)*100:.1f}%)")
print(f"  MLB sharpness drop: {sharpness_drop_mlb:.1f}%")
print(f"  MSV sharpness drop: {sharpness_drop_msv:.1f}%")

if brightness_drop_mlb > 20 or brightness_drop_msv > 20:
    print(f"  Noisy images are significantly darker")
if sharpness_drop_mlb > 20 or sharpness_drop_msv > 20:
    print(f"  Noisy images are significantly blurrier - good for testing preprocessing!")

# Save metrics for later reference in preprocessing
SHARPNESS_DROP_MLB = sharpness_drop_mlb
SHARPNESS_DROP_MSV = sharpness_drop_msv
BRIGHTNESS_DROP_MLB = brightness_drop_mlb
BRIGHTNESS_DROP_MSV = brightness_drop_msv

print("\nMetrics saved for preprocessing justification.")

    #Quality analysis revealed that noisy images had similar brightness to clean data (118.5 vs 118.9) but suffered severe sharpness degradation, with MLB images losing 81.1% sharpness and MSV images losing 59.8% sharpness. This guided our preprocessing strategy to focus on sharpening techniques rather than brightness correction.
```

```python
# CELL: Sharpening Function Definition

def sharpen_image(image, kernel_size=5, sigma=1.0, amount=1.5, threshold=0):
    """
    Sharpen image using unsharp mask technique to address blur in noisy images.
    
    Parameters:
    - kernel_size: Size of Gaussian blur kernel (default: 5)
    - sigma: Gaussian blur standard deviation (default: 1.0)
    - amount: Sharpening strength, 1.0-2.0 (default: 1.5)
    - threshold: Minimum brightness difference to sharpen (default: 0)
    
    Returns:
    - Sharpened image (numpy array)
    
    Method: Unsharp Masking
    Formula: sharpened = original + amount * (original - blurred)
    """
    blurred = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)
    sharpened = cv2.addWeighted(image, 1.0 + amount, blurred, -amount, 0)
    
    if threshold > 0:
        low_contrast_mask = np.absolute(image - blurred) < threshold
        np.copyto(sharpened, image, where=low_contrast_mask)
    
    return sharpened

print("Sharpening function defined.")
```

```python
# CELL: Preprocessing Pipeline

def preprocess_image(image, kernel_size=5, sigma=1.0, amount=1.5):
    """
    Preprocessing pipeline for noisy greenhouse images
    
    Applies sharpening to address blur degradation identified in quality analysis.
    Based on findings showing 76% sharpness loss in MLB and 60% in MSV images.
    
    Parameters:
    - image: Input image (BGR format from cv2.imread)
    - kernel_size: Gaussian blur kernel size for unsharp mask (default: 5)
    - sigma: Gaussian blur standard deviation (default: 1.0)
    - amount: Sharpening strength (default: 1.5)
    
    Returns:
    - Preprocessed (sharpened) image
    """
    processed = sharpen_image(image, kernel_size=kernel_size, sigma=sigma, amount=amount)
    return processed

print("Preprocessing pipeline defined")
print("\nPipeline: Unsharp mask sharpening")
print(f"  Targets blur problem (MLB: {SHARPNESS_DROP_MLB:.1f}% loss, MSV: {SHARPNESS_DROP_MSV:.1f}% loss)")
print("  Brightness correction not applied (minimal difference observed)")
```

```python
# CELL: Batch Process All Test Images

import time

# Create output directory for preprocessed images
PREPROCESSED_DIR = PROCESSED_DIR / "test_preprocessed"
PREPROCESSED_DIR.mkdir(exist_ok=True)
(PREPROCESSED_DIR / "MLB").mkdir(exist_ok=True)
(PREPROCESSED_DIR / "MSV").mkdir(exist_ok=True)

print("Batch processing all test images with sharpening...")
print("="*70)

# Process MLB images
MLB_DIR = PROCESSED_DIR / "test" / "MLB"
print("\nProcessing MLB images...")
mlb_images = list(MLB_DIR.glob("*.jpg"))
start_time = time.time()

for i, img_path in enumerate(mlb_images):
    img = cv2.imread(str(img_path))
    if img is not None:
        # Apply preprocessing (sharpening only)
        preprocessed = preprocess_image(img)
        
        # Save to preprocessed folder
        output_path = PREPROCESSED_DIR / "MLB" / img_path.name
        cv2.imwrite(str(output_path), preprocessed)
    
    if (i + 1) % 200 == 0:
        print(f"  Processed {i + 1}/{len(mlb_images)} images...")

mlb_time = time.time() - start_time
print(f"  Completed {len(mlb_images)} MLB images in {mlb_time:.1f} seconds")

# Process MSV images
MSV_DIR = PROCESSED_DIR / "test" / "MSV"
print("\nProcessing MSV images...")
msv_images = list(MSV_DIR.glob("*.jpg"))
start_time = time.time()

for i, img_path in enumerate(msv_images):
    img = cv2.imread(str(img_path))
    if img is not None:
        # Apply preprocessing (sharpening only)
        preprocessed = preprocess_image(img)
        
        # Save to preprocessed folder
        output_path = PREPROCESSED_DIR / "MSV" / img_path.name
        cv2.imwrite(str(output_path), preprocessed)
    
    if (i + 1) % 200 == 0:
        print(f"  Processed {i + 1}/{len(msv_images)} images...")

msv_time = time.time() - start_time
print(f"  Completed {len(msv_images)} MSV images in {msv_time:.1f} seconds")

print("\n" + "="*70)
print("BATCH PROCESSING COMPLETE")
print("="*70)
print(f"Total images processed: {len(mlb_images) + len(msv_images)}")
print(f"Total processing time: {mlb_time + msv_time:.1f} seconds")
print(f"Average time per image: {(mlb_time + msv_time)/(len(mlb_images) + len(msv_images)):.3f} seconds")
print(f"\nPreprocessed images saved to: {PREPROCESSED_DIR}")
print("  Structure: test_preprocessed/MLB/ and test_preprocessed/MSV/")
print("="*70)
```

```python
## 6. Model Training

### 6.1 Training MobileNetV2 on Clean Data

#We train MobileNetV2 using transfer learning on the PlantVillage dataset (33,013 training images, 4,126 validation images). The model uses ImageNet pre-trained weights and is fine-tuned for plant disease classification.
```

```python
# CELL: Train MobileNetV2

import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define MODEL_PATH first
MODEL_PATH = BASE_DIR / "models"
MODEL_PATH.mkdir(exist_ok=True)

# Check if model already exists
if (MODEL_PATH / "mobilenetv2_plantvillage.h5").exists():
    print("="*70)
    print("MODEL ALREADY TRAINED")
    print("="*70)
    print(f"Existing model found: {MODEL_PATH / 'mobilenetv2_plantvillage.h5'}")
    print("Skipping training to save time (~20 minutes).")
    print("\nTo retrain from scratch, delete the model file first.")
    print("="*70)
else:
    print("Setting up MobileNetV2 for training...")
    print("="*70)
    
    # Image parameters
    IMG_SIZE = 224
    BATCH_SIZE = 32
    EPOCHS = 20
    
    # Data paths
    TRAIN_DIR = PROCESSED_DIR / "train"
    VAL_DIR = PROCESSED_DIR / "val"
    
    # Count number of classes
    num_classes = len([d for d in TRAIN_DIR.iterdir() if d.is_dir()])
    print(f"Number of classes: {num_classes}")
    print(f"Image size: {IMG_SIZE}x{IMG_SIZE}")
    print(f"Batch size: {BATCH_SIZE}")
    print(f"Maximum epochs: {EPOCHS}")
    
    # Data augmentation for training
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        zoom_range=0.2,
        fill_mode='nearest'
    )
    
    val_datagen = ImageDataGenerator(rescale=1./255)
    
    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=True
    )
    
    val_generator = val_datagen.flow_from_directory(
        VAL_DIR,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )
    
    print(f"\nTraining samples: {train_generator.samples}")
    print(f"Validation samples: {val_generator.samples}")
    
    # Build model
    print("\nBuilding MobileNetV2 model...")
    
    base_model = MobileNetV2(
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
        include_top=False,
        weights='imagenet'
    )
    
    base_model.trainable = False
    
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.2)(x)
    predictions = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    print("Model compiled successfully")
    print(f"Total parameters: {model.count_params():,}")
    
    # Callbacks
    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        verbose=1
    )
    
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
    
    # Train
    print("\n" + "="*70)
    print("Starting training...")
    print("="*70 + "\n")
    
    history = model.fit(
        train_generator,
        epochs=EPOCHS,
        validation_data=val_generator,
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )
    
    print("\n" + "="*70)
    print("Training complete!")
    print("="*70)
    
    # Save model
    model.save(MODEL_PATH / "mobilenetv2_plantvillage.h5")
    print(f"\nModel saved to: {MODEL_PATH / 'mobilenetv2_plantvillage.h5'}")
```

```python
### 7.2 Testing on Noisy Images (Without Preprocessing)

#We now test the same model on real-world noisy greenhouse images to measure performance degradation.
```

```python
# CELL: Test on Noisy Images (No Preprocessing)

# We need to create a generator for noisy images
# But noisy images only have maize and tomato, not all 15 classes
# So we'll test differently - load images manually and predict

from tensorflow.keras.preprocessing import image as keras_image

print("="*70)
print("TESTING ON NOISY IMAGES (Without Preprocessing)")
print("="*70)

# Get all noisy images
noisy_maize = list(MAIZE_DIR.rglob("*.jpg"))
noisy_tomato = list(TOMATO_DIR.rglob("*.jpg"))
all_noisy = noisy_maize + noisy_tomato

print(f"\nTotal noisy images: {len(all_noisy)}")
print(f"  Maize: {len(noisy_maize)}")
print(f"  Tomato: {len(noisy_tomato)}\n")

# We'll predict and see how many it gets right
# Since these are just "diseased" or "healthy" leaves, we'll check if predictions are reasonable

predictions = []
print("Making predictions on noisy images...")

for i, img_path in enumerate(all_noisy):
    # Load and preprocess image
    img = keras_image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = keras_image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0
    
    # Predict
    pred = model.predict(img_array, verbose=0)
    pred_class = np.argmax(pred)
    pred_confidence = np.max(pred)
    
    predictions.append({
        'image': img_path.name,
        'predicted_class': class_names[pred_class],
        'confidence': pred_confidence
    })
    
    if (i + 1) % 500 == 0:
        print(f"  Processed {i + 1}/{len(all_noisy)} images...")

print(f"  Completed all {len(all_noisy)} images\n")

# Analyze predictions
avg_confidence = np.mean([p['confidence'] for p in predictions])
low_confidence = sum(1 for p in predictions if p['confidence'] < 0.5)

print("="*70)
print("NOISY TEST RESULTS (Without Preprocessing)")
print("="*70)
print(f"Average confidence: {avg_confidence:.4f} ({avg_confidence*100:.2f}%)")
print(f"Low confidence predictions (< 50%): {low_confidence} ({low_confidence/len(all_noisy)*100:.1f}%)")

# Show distribution of predicted classes
from collections import Counter
pred_classes = [p['predicted_class'] for p in predictions]
class_dist = Counter(pred_classes)

print("\nPredicted class distribution:")
for cls, count in class_dist.most_common(10):
    print(f"  {cls:50s} {count:4d} ({count/len(all_noisy)*100:.1f}%)")

print("="*70)
```

```python
### 7.3 Testing on Noisy Images (With Preprocessing)

#We now test on the same noisy images after applying our preprocessing pipeline (brightness correction + sharpening) to evaluate if preprocessing improves model performance.
```

```python
# CELL: Test on Noisy Images (With Preprocessing)

print("="*70)
print("TESTING ON NOISY IMAGES (With Preprocessing)")
print("="*70)

# Get all preprocessed noisy images
preprocessed_maize = list((PREPROCESSED_DIR / "maize").rglob("*.jpg"))
preprocessed_tomato = list((PREPROCESSED_DIR / "tomato").rglob("*.jpg"))
all_preprocessed = preprocessed_maize + preprocessed_tomato

print(f"\nTotal preprocessed images: {len(all_preprocessed)}")
print(f"  Maize: {len(preprocessed_maize)}")
print(f"  Tomato: {len(preprocessed_tomato)}\n")

predictions_preprocessed = []
print("Making predictions on preprocessed noisy images...")

for i, img_path in enumerate(all_preprocessed):
    # Load and preprocess image
    img = keras_image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = keras_image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0
    
    # Predict
    pred = model.predict(img_array, verbose=0)
    pred_class = np.argmax(pred)
    pred_confidence = np.max(pred)
    
    predictions_preprocessed.append({
        'image': img_path.name,
        'predicted_class': class_names[pred_class],
        'confidence': pred_confidence
    })
    
    if (i + 1) % 500 == 0:
        print(f"  Processed {i + 1}/{len(all_preprocessed)} images...")

print(f"  Completed all {len(all_preprocessed)} images\n")

# Analyze predictions
avg_confidence_prep = np.mean([p['confidence'] for p in predictions_preprocessed])
low_confidence_prep = sum(1 for p in predictions_preprocessed if p['confidence'] < 0.5)

print("="*70)
print("NOISY TEST RESULTS (With Preprocessing)")
print("="*70)
print(f"Average confidence: {avg_confidence_prep:.4f} ({avg_confidence_prep*100:.2f}%)")
print(f"Low confidence predictions (< 50%): {low_confidence_prep} ({low_confidence_prep/len(all_preprocessed)*100:.1f}%)")

# Show distribution
pred_classes_prep = [p['predicted_class'] for p in predictions_preprocessed]
class_dist_prep = Counter(pred_classes_prep)

print("\nPredicted class distribution:")
for cls, count in class_dist_prep.most_common(10):
    print(f"  {cls:50s} {count:4d} ({count/len(all_preprocessed)*100:.1f}%)")

print("="*70)

# COMPARISON
print("\n" + "="*70)
print("COMPARISON: ORIGINAL vs PREPROCESSED NOISY IMAGES")
print("="*70)
print(f"{'Metric':<40} {'Original':<15} {'Preprocessed':<15} {'Change'}")
print("-"*70)
print(f"{'Average Confidence':<40} {avg_confidence*100:>6.2f}%        {avg_confidence_prep*100:>6.2f}%        {(avg_confidence_prep-avg_confidence)*100:+.2f}%")
print(f"{'Low Confidence (< 50%)':<40} {low_confidence:>6d}         {low_confidence_prep:>6d}         {low_confidence_prep-low_confidence:+d}")
print("="*70)
```
